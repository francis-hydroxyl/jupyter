```python
data_fits_R = glob.glob(f'/home/elmegreen/jupyter/research/Bubble_detection/Paper2/phangs_jwst/Detection_Model/SN_Model_New_Norm/analyse_JWST/noise/smooth_fits/{gal_name}/*f2100w*0.25.smooth0.7arcsec.fits')[0] ##2D
data_fits_G = glob.glob(f'/home/elmegreen/jupyter/research/Bubble_detection/Paper2/phangs_jwst/Detection_Model/SN_Model_New_Norm/analyse_JWST/noise/smooth_fits/{gal_name}/*f770w*0.1.smooth0.3arcsec.fits')[0]##2D
data_fits_B = glob.glob(f'/home/elmegreen/jupyter/research/Bubble_detection/Paper2/phangs_jwst/Detection_Model/SN_Model_New_Norm/analyse_JWST/noise/smooth_fits/{gal_name}/*f770w*0.1.smooth0.3arcsec.fits')[0]
```

```python
hdu_r = astropy.io.fits.open(data_fits_R)[0]
hdu_g = astropy.io.fits.open(data_fits_G)[0]
hdu_b = astropy.io.fits.open(data_fits_B)[0]
```

```python
print(np.nanmax(hdu_r.data), np.nanmin(hdu_r.data), np.nanmean(hdu_r.data))
print(np.nanmax(hdu_g.data), np.nanmin(hdu_g.data), np.nanmean(hdu_g.data))
```

```python
color_min_R = 0.8
color_max_R = 2.3
color_min_G = 0.2
color_max_G = 1.7
# color_min_G = 1000
# color_max_G = 1000
color_min_B = 1000
color_max_B = 1000
```

```python
colorval = "%.1f_%.1f_%.1f_%.1f_%.1f_%.1f"%(color_min_R, color_max_R, color_min_G, color_max_G, color_min_B, color_max_B)
save_png_name = 'RGB_fig_APLPY/ic5332_RG_%s'%colorval + '_smooth.png'
fitss = [data_fits_R, data_fits_G, data_fits_B]
```

```python
aplpy.make_rgb_image(fitss, save_png_name, 
                     vmin_r=color_min_R, 
                     vmax_r=color_max_R,
                     vmin_g=color_min_G,
                     vmax_g=color_max_G, 
                     vmin_b=color_min_B, 
                     vmax_b=color_max_B, 
                     stretch_r="power", 
                     # stretch_r="log", 
                     stretch_g="sqrt", 
                     # stretch_g="log", 
                     stretch_b="linear")
```

```python
fig = plt.figure(figsize=(16, 16))
f = aplpy.FITSFigure(data_fits_G, slices=[0], figure=fig, convention='wells')#, north=True)
f.show_rgb(save_png_name)
```

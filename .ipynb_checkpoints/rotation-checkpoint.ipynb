```python
import numpy as np
import cv2
import os,glob,random
```

```python
save_dir="/home/filament/jupyter/Paper/each_picture/Rank1_rot/"  
files=glob.glob("/home/filament/jupyter/Paper/each_picture/Rank1/*.png")
n=0
for l in files:
    img=cv2.imread(l)
    outfile=save_dir+"/"+str(n)+".jpg"
    #　反時計回りに回転させる
    img=cv2.rotate(img , cv2.ROTATE_90_COUNTERCLOCKWISE)
    cv2.imwrite(outfile, img)
    n+=1
```

```python
import astroquery.vizier
import pandas as pd
import numpy as np
```

```python
viz = astroquery.vizier.Vizier(columns=["*"])
viz.ROW_LIMIT = -1
bub_2006 = viz.query_constraints(catalog="J/ApJ/649/759/bubbles")[0].to_pandas()
bub_2007 = viz.query_constraints(catalog="J/ApJ/670/428/bubble")[0].to_pandas()
bub_2006_change = bub_2006.set_index("[CPA2006]")
bub_2007_change = bub_2007.set_index("[CWP2007]")
CH = pd.concat([bub_2006_change, bub_2007_change])
CH["CH"] = CH.index
CH_name = np.load('/home/filament/galactic_bubble/photoutils_clustering/rank_3.npy')
CH = CH.loc[CH_name]
CH
```

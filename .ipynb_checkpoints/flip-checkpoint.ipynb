```python
import numpy as np
import cv2
import os,glob,random
```

```python
save_dir = "/home/filament/jupyter/Paper/each_picture/flip/right_left/Rank1/"
files = glob.glob("/home/filament/jupyter/Paper/each_picture/Rank1/*.png")

for l in files:
    original_filename = os.path.basename(l)
    basename, ext = os.path.splitext(original_filename)
    new_filename = f"{basename}_flip_right_left_{ext}"
    outfile = os.path.join(save_dir, new_filename)
    img = cv2.imread(l)
    # 左右回転させる
    img = cv2.flip(img, 1)
    cv2.imwrite(outfile, img)
```

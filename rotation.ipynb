```python
import numpy as np
import cv2
import os,glob,random
```

```python
save_dir = "/home/filament/jupyter/Paper/each_picture/rot/Rank1/rot_90_counterclock/"
files = glob.glob("/home/filament/jupyter/Paper/each_picture/Rank1/*.png")

for l in files:
    original_filename = os.path.basename(l)
    basename, ext = os.path.splitext(original_filename)
    new_filename = f"{basename}_rot_90_counterclock{ext}"
    outfile = os.path.join(save_dir, new_filename)
    img = cv2.imread(l)
    # 反時計回りに回転させる
    img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    cv2.imwrite(outfile, img)
```
